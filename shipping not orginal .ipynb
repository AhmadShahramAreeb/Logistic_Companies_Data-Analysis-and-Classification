## 1. Random Forest  Classifier
* Random Forest Classifier, danışmanlı öğrenme algoritmasıyla çalışan, sınıflandırma ve regresyon problemleri için kullanılan, birden fazla karar ağacının (ensemble) bir araya gelmesiyle oluşturulan bir sınıflandırma yöntemidir.r.

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=100,random_state=42)

#  sklearn kütüphanesinden RandomForestClassifier sınıfını içe aktarır ve bir RandomForestClassifier nesnesi oluşturur.
# n_estimators parametresi, oluşturulacak olan karar ağacı sayısını belirler
# random_state, modelin rastgeleliğini kontrol eden bir parametredir
```

```python
RF.fit(x_train,y_train)
# Bu kod, RandomForestClassifier modelini eğitmek için kullanılır. Eğitim veri seti (x_train) ve hedef değişken (y_train) kullanılarak model eğitilir.
# Model, veri setindeki desenleri öğrenir ve daha sonra bu desenleri kullanarak yeni verilere tahmin yapabilir.
```

```python
y_pred_rf=RF.predict(x_train)
y_pred_rf

# Bu kod, eğitilmiş RandomForestClassifier modelini kullanarak eğitim veri seti üzerinde tahminler yapar ve bu tahminleri y_pred_rf değişkenine atar.
```

```python
accuracy=accuracy_score(y_pred_rf,y_train)
print(f"Accuracy: {accuracy}")
# Bu kod, tahmin edilen sınıflar ile gerçek sınıflar arasındaki doğruluğu hesaplar. Bunu yapmak için accuracy_score fonksiyonu kullanılır.
```

```python
pred_rf=RF.predict(x_test)
# Bu kod, eğitilmiş RandomForestClassifier modelini kullanarak test veri seti üzerinde tahminler yapar.

# Yapılan tahminler, modelin test veri setindeki örnekleri nasıl sınıflandırdığını gösterir.
```

## Doğruluk Oranı

```python
accuracy=accuracy_score(pred_rf,y_test)
print(f"accuracy:{accuracy}")

# Bu kod, RandomForestClassifier modelinin test veri seti üzerindeki doğruluğunu hesaplar.
```

```python
report2 = classification_report(y_test, pred_rf)
print("Classification Report:")
print(report2)
# Bu kod, sklearn kütüphanesinin classification_report fonksiyonunu kullanarak bir sınıflandırma raporu oluşturur.
# Bu rapor, bir modelin sınıflandırma performansını ölçmek için yaygın olarak kullanılan bir matris setini sağlar. 
# Rapor, sınıf bazında doğruluk, hassasiyet, geri çağırma ve F1 skoru gibi değerleri içerir.

# Precision (Hassasiyet):H =(TP/(TP+FP)) Hassasiyet, doğru pozitif tahminlerin toplam pozitif tahminler içindeki oranını ifade eder. 
# Yani, gerçekte pozitif olanların ne kadarının doğru bir şekilde tahmin edildiğini gösterir. 
# Hassasiyet, yanlış pozitiflerin minimize edilmesi gereken durumlarda önemlidir.

# Recall (Duyarlılık veya Geri Çağırma): R.C =(TP/(TP+FN)) Duyarlılık, gerçek pozitiflerin tüm pozitifler içindeki oranını ifade eder. 
# Yani, gerçek pozitiflerin ne kadarının doğru bir şekilde tahmin edildiğini gösterir. Duyarlılık, yanlış negatiflerin minimize edilmesi gereken durumlarda önemlidir.

# F1-Score: F1=(2*(H*R.C/(H+R.C))) F1 skoru, hassasiyet ve duyarlılık arasındaki dengeyi gösteren bir ölçüdür. 
# F1 skoru, hassasiyet ve duyarlılığın harmonik ortalamasını alarak hesaplanır.

# Support (Destek): Destek, her bir sınıf için gerçek etiketlerin toplam sayısını ifade eder.

# Accuracy (Doğruluk): A=(Tp/toplam tahmin) Doğruluk, doğru tahmin edilen örneklerin toplam örnek sayısına oranını ifade eder.

# Macro Average (Makro Ortalama): Makro ortalama, sınıf başına hesaplanan değerlerin aritmetik ortalamasını alır. Her sınıfın katkısı eşit olarak değerlendirilir.

# Weighted Average (Ağırlıklı Ortalama): Ağırlıklı ortalama, her sınıfın destek sayısıyla ağırlıklı olarak hesaplanan değerlerin ortalamasını alır. Büyük sınıfların daha fazla ağırlığı vardır.

# Support (Destek): Destek, her bir sınıf için gerçek etiketlerin toplam sayısını ifade eder.
```

## 2. GradientBoostingClassifier

* Zayıf öğreniciler (weak learners), tek başına yüksek doğruluk sağlayamayan ancak bir araya geldiklerinde güçlü bir tahmin modeli oluşturabilen basit modellerdir

```python
from sklearn.ensemble import GradientBoostingClassifier
GBC=GradientBoostingClassifier(loss='log_loss',max_depth=3,
    min_samples_split=2,
    min_samples_leaf=1,
    learning_rate=0.1,
    n_estimators=100,random_state=42)
GBC.fit(x_train,y_train)
```

* sklearn kütüphanesinden GradientBoostingClassifier sınıfını kullanarak bir

```python
y_pred_gb=GBC.predict(x_train)
y_pred_gb
```

* Bu kod, eğitilmiş Gradient Boosting Classifier modelini kullanarak eğitim veri seti üzerinde tahminler yapar
* bu tahminleri y_pred_gb değişkenine atar

```python
accuracy=accuracy_score(y_pred_gb,y_train)
print(f"Accuracy: {accuracy}")
```

* Gradient Boosting Classifier modelinin eğitim veri seti üzerindeki doğruluğunu hesaplar ve sonucu ekrana yazdırır.
